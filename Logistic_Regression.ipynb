```python
def costFunction(X,y,Theta):
    e = 0.00001
    z = np.dot(X,Theta)
    h = sigmoid(z)
    J = np.sum(-y*np.log(h+e)-(1-y)*np.log(1-h+e))/m
    return [h,J]
```

```python
# Gradient Descent(X,y,Theta) returns Theta
def gradientDescent(X,y,Theta,alpha):
    noOfParam = Theta.shape[0]
    h = costFunction(X,y,Theta)[0]
    J = costFunction(X,y,Theta)[1]
    for i in range(noOfParam):
        Theta[i,0] = Theta[i,0]-alpha*np.sum((h-y)*X[:,i].reshape(m,1))/m
    #Theta = Theta - np.dot(X.transpose(),(h-y))
    return [h,J,Theta]
```

```python
def learningLogisticRegression(X,y,alpha,noOfIter,printIter):
    jHistory= []
    e = 0.0000001
    Theta = np.zeros((X.shape[1],1))
    for i in range(noOfIter):
        h = 1/(1+np.exp(-np.dot(X,Theta)))
        J = np.sum(-y*np.log(h+e)-(1-y)*np.log(1-h+e))/m
        Theta = Theta - alpha*np.dot(X.transpose(),(h-y))/m
        jHistory.append(J)
        if (i%printIter==0):
            print('Iteration = ',i)
            print('Cost = ',J)
            print('Theta = ', Theta)

    plt.plot(list(range(noOfIter)),jHistory)
    return Theta
```
